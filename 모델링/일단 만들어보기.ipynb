```python
# y값 추출
y_train_text = [ i.split('\t')[-1] for i in train_text.split('\n') ]

# y 전처리
import re

#개행문자 \r 제거
y_train_text = [ re.sub('\r','',s) for s in y_train_text]

#점수가없는 공백열 제거

#문자형을 숫자형으로 변환
y_train_text = [int(i) for i in y_train_text if (len(i)<=2 and len(i)!=0)]  #
y_train_text
```
